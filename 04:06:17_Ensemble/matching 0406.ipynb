```python
import pandas as pd
import numpy as np
```

```python
idf = pd.read_excel("ios_matching_input.xlsx") 
idf.head()
```

```python
idf['app_name'] = idf['app_name'].str.lower()
```

```python
import string
def pre(row):
    s = row['app_name']
    lst = [ch for ch in s if ch not in string.punctuation]
    
    new_s = ''.join(lst)
    new_s = new_s.replace('  ',' ')
    row['new_app_name'] = new_s
    return row
```

```python
idf = idf.apply(pre,axis =1)
```

```python
idict = idf['new_app_name'].to_dict()
```

```python
adf = pd.read_excel("Andriod_matching_input.xlsx")
adf.head()
```

```python
adf['name'] = adf['name'].str.lower()
```

```python
def pre(row):
    s = str(row['name'])
    lst = [ch for ch in s if ch not in string.punctuation]
    
    new_s = ''.join(lst)
    new_s = new_s.replace('  ',' ').replace('free','')
    new_s.strip()
    row['new_name'] = new_s
    return row
```

```python
adf= adf.apply(pre,axis=1)
```

```python
##1. strict match == 
def find(row):
    for k,v in idict.items():
        if str(row['name'])==str(v):
            row['i_idx'] = k
            row['i_level'] = idf.ix[k,'rating']
    return row
```

```python
adf = adf.apply(find,axis=1)
```

```python
len(adf[adf['i_idx']!=np.nan])
```

```python
###2. contains 
def find2(row):
    for k,v in idict.items():
        if adf[adf['i_idx']==np.nan:
            if str(row['name']) in str(v):
                row['i_idx'] = k
                row['i_level'] = idf.ix[k,'rating']
    return row
```

```python
adf = adf.apply(find2,axis=1)
```

```python
##3. pandas match 
def find3(row):
    try:
        boo = idf['app_name'].str.contains(str(row['name']))
        idx = boo[boo==True].index[0]
        row['i_idx']= idx
        row['i_level'] = idf.ix[idx,'rating']
    except:
        pass
    
    return row
```

```python
adf = adf.apply(find3,axis=1)
```
