```python
import nltk
from nltk.corpus import stopwords
import string
import os
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Violence = open('Violence.txt','r').read().replace('\n',' ')
Sex = open('Sex.txt','r').read().replace('\n',' ')
Language = open('Language.txt','r').read().replace('\n',' ')
ADT = open('ADT.txt','r').read().replace('\n',' ')
Gambling = open('Gambling.txt','r').read().replace('\n',' ')
```

```python
def process(input_text):
    #Tokenization
    tokens = nltk.word_tokenize(input_text)
    #Remove stop words
    mystopwords = stopwords.words('english')
    words = [w.lower() for w in tokens if w.isalpha() if w.lower()not in mystopwords]
    
    freq = nltk.FreqDist(words)
    
    return freq
    
V_freq = process(Violence)
S_freq = process(Sex)
L_freq = process(Language)
A_freq = process(ADT)
G_freq = process(Gambling)
```

```python
Total_freq = V_freq+S_freq+L_freq+A_freq+G_freq
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(Total_freq,orient = 'index')
```

```python
dfV = pd.DataFrame.from_dict(V_freq,orient = 'index')
dfV.columns = ['V_freq']
dfS = pd.DataFrame.from_dict(S_freq,orient = 'index')
dfS.columns = ['S_freq']
dfL = pd.DataFrame.from_dict(L_freq,orient = 'index')
dfL.columns = ['L_freq']
dfA = pd.DataFrame.from_dict(A_freq,orient = 'index')
dfA.columns = ['A_freq']
dfG = pd.DataFrame.from_dict(G_freq,orient = 'index')
dfG.columns = ['G_freq']
```

```python
df = pd.merge(df,dfV,how = 'left',left_index = True, right_index = True)
```

```python
df = pd.merge(df,dfS,how = 'left',left_index = True, right_index = True)
df = pd.merge(df,dfL,how = 'left',left_index = True, right_index = True)
df = pd.merge(df,dfA,how = 'left',left_index = True, right_index = True)
df = pd.merge(df,dfG,how = 'left',left_index = True, right_index = True)
```

```python
df
```

```python
df = df.fillna(0)
```

```python
#Normalize

df['V_freq'] = df['V_freq']/sum(df['V_freq'])
df['S_freq'] = df['S_freq']/sum(df['S_freq'])
df['L_freq'] = df['L_freq']/sum(df['L_freq'])
df['A_freq'] = df['A_freq']/sum(df['A_freq'])
df['G_freq'] = df['G_freq']/sum(df['G_freq'])
```

```python
df_norm = df.ix[:,1:6]*1000000
df_norm.columns = ['V_freq_norm','S_freq_norm','L_freq_norm','A_freq_norm','G_freq_norm']
```

```python
df_norm
```

```python
result = pd.merge(df,df_norm,how='outer',left_index=True, right_index = True)
```
