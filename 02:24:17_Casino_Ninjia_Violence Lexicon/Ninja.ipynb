```python
import nltk
import pandas as pd
from nltk import FreqDist
import numpy as np
import string
```

```python
FV = open('Fantasy_Violence.txt','r').read()
V = open('Violence.txt','r').read()
B = open('Blood','r').read()
```

```python
porter = nltk.PorterStemmer()
```

```python
def process(txt):
    txt = ''.join([ch for ch in txt if ch not in string.punctuation])
        
    tokens = nltk.word_tokenize(txt)
    tokens = [w.lower() for w in tokens]
    tags = nltk.pos_tag(tokens)
    ##only keep n, v, adj, adv
    tag_list = [(word,tag) for (word,tag) in tags if tag.startswith('N') or tag.startswith('V') or tag.startswith('J') or tag.startswith('R')]
    
    ## Frequency distribution ##
    tag_freq = FreqDist(tag_list)
    
    sorted_tag_freq = sorted(tag_freq.items(), key = lambda k:k[1], reverse = True)
    
    return sorted_tag_freq    
```

```python
FV_freq = process(FV)
FV_freq
```

```python
V_freq = process(V)
V_freq
```

```python
B_freq = process(B)
B_freq
```

```python
Ninja = pd.read_csv('Ninja.csv')
Ninja = Ninja[['name','content_rating','content_rating_reason','description']]
Ninja.head()
```

```python
description = Ninja.ix[6,'description']
description = ''.join([ch for ch in description if ch not in string.punctuation])
des_list = description.split(' ')

tokens = nltk.word_tokenize(' '.join(des_list))
tokens = [w for w in tokens if w.isalpha()]
```

```python
description
```

```python
def processrow(row):
    description = row['description']
    description = ''.join([ch for ch in description if ch not in string.punctuation])
    des_list = description.split(' ')
    
    tokens = nltk.word_tokenize(' '.join(des_list))
    tokens = [w for w in tokens if w.isalpha()]
    row['Word_Count'] = len(tokens)
    freq = FreqDist(tokens) #dictionary(word:count)
    
    V1_freq=0
    V2_freq=0
    V3_freq=0
    
    for word,count in freq.items():
        if word in V1:
            V1_freq += count
        elif word in V2:
            V2_freq += count
        elif word in V3:
            V3_freq += count
    
    row['V1_freq'] = V1_freq
    row['V2_freq'] = V2_freq
    row['V3_freq'] = V3_freq
    
    return row
```

```python
Ninja = Ninja.apply(processrow,axis = 1)
```

```python
Ninja.to_csv('Ninja_V1V2V3.csv') 
```
